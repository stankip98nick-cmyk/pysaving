```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
from sklearn.linear_model import LinearRegression
import joblib
```

```python
df = pd.read_csv("sales.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
x = df[['Marketing_Spend','Store_Size','Month_Number']]
y = df['Sales']
```

```python
print(x.shape)
```

```python
print(y.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
my_model = LinearRegression()
my_model.fit(x_train,y_train)
```

```python
y_pred = my_model.predict(x_test)
y_pred
```

```python
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
from math import sqrt
```

```python
rmse = sqrt(mse)
rmse
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
from fastapi import FastAPI
from pydantic import BaseModel
import joblib
import numpy as np
```

```python
my_model=joblib.load('sales.joblib')
feature_names = joblib.load('sales_features.joblib')
```

```python
sales = FastAPI(title='sales.API')
```

```python
class SalesFeatures(BaseModel):
    Marketing_Spend: float
    Store_Size:float
    Month_Number:int
```

```python
@ sales.get('/')
def root():
    return{'message':'sales api online!!!'}
```

```python
@ sales.post('/')
def predict_sales(data:SalesFeatures):
    x = np.array([getattr(data,f) for f in feature_names ])
    prediction = my_model.predict(x)[0]
    return{'predict_sales':float(prediction)}
```
